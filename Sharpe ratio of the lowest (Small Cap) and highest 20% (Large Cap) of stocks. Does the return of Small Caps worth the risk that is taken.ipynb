```python
returns=pd.read_csv("data/portfolios_formed_on_ME_monthly_EW.csv", index_col=[0], parse_dates=True)
returns.head()
```

```python
columns=["Lo 20", "Hi 20"]
returns=returns[columns]
returns.columns=["Small Cap", "Large Cap"]
returns.head()
```

```python
riskfree_rate=0.03
annualized_vol=returns.std()*np.sqrt(12)
n_months=returns.shape[0]
annualized_vol
```

```python
annualized_return=((returns/100)+1).prod()**(12/n_months)-1
annualized_return
```

```python
sharpe=(annualized_return-riskfree_rate)/annualized_vol
sharpe
```
