```python
#Hello. Below, I analyse csv dataset in which I have stocks between 1926-2018.I will calculate Sharpe ratio to find if it is worth to invest in Small Cap stocks considering the risk that they bear. 

#importing pandas and numpy for calculations
import pandas as pd
import numpy as np
```

```python
#I have imported the csv file as returns and showed first 5 lines
returns=pd.read_csv("data/portfolios_formed_on_ME_monthly_EW.csv", index_col=[0], parse_dates=True)
returns.head()
```

```python
#As I analyse low 20% and high 20% market cap, I narrowed down columns to leave only these two
columns=["Lo 20", "Hi 20"]
returns=returns[columns]
#renamed columns to Small Cap (low 20%) and Large cap (high 20%)
returns.columns=["Small Cap", "Large Cap"]
returns.head()
```

```python
#sharpe ratio=(annualized return-riskfree rate)/annualized volatility
#risk free rate is the rate that usually US T bonds have. On average, those are 3%
riskfree_rate=0.03
#volatility is the ratio which describes how much stock varies from the mean return
#to find annualized volatility standard deviation of return is multiplied by square root of 12 months, as we have 12 months in a year.
annualized_vol=returns.std()*np.sqrt(12)
#in the csv dataset all the date is described in months, thus we need an array to find how many months we have there
n_months=returns.shape[0] #this returns 1110 rows which is 1110 months in total from 1926 to 2018
annualized_vol
```

```python
annualized_return=((returns/100)+1).prod()**(12/n_months)-1 #I have modifed returns in 1+r format
annualized_return
```

```python
Sharpe_ratio=(annualized_return-riskfree_rate)/annualized_vol
Sharpe_ratio
```
